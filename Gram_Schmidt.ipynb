```python
import numpy as np

#gram-schmidt process automated using numpy
```

```python

#manually input the vectors in U

def gram_schmidt(vectors):
    orthonormal_basis = []

    for v in vectors:                 #loops through every vector in U
        v = v.astype(float)

        for u in orthonormal_basis:
            v -= np.dot(v, u) * u     #projection

        v = v / np.linalg.norm(v)      #normalises vector

        orthonormal_basis.append(v)   #adds to list

    return np.array(orthonormal_basis)


U = np.array(
    [[1,2,3],
     [0,1,4],
     [5,6,0]]
)

print(f'Input:\n {U}')
print(f'Output: {gram_schmidt(U)}')
```
