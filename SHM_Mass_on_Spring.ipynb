```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
m = 1
k = 5
omega2 = -k/m
c = 0.6

#non damped system
def system(t,y):
    x, v = y
    dxdt = v
    dvdt = (-k/m) *x
    return [dxdt, dvdt]

y0 = [2, 0]

t_span = (0,10)
t_eval = np.linspace(0,10,1000)

solution = solve_ivp(system, t_span, y0, t_eval=t_eval)



#damped system

def system_spring(t,y):
    x, v = y
    dxdt = v
    dvdt = ((-k/m)*x) -((c/m)*v)
    return [dxdt, dvdt]



solutionset = solve_ivp(system_spring, t_span, y0, t_eval = t_eval)




print(solution)
print(solutionset)
```

```python
#Plot non damped - displacement and velocity
plt.plot(solution.t, solution.y[0], color = 'black', label='Displacement')

plt.plot(solution.t, solution.y[1], color = 'blue', label='Velocity')

plt.xlabel('Time')
plt.grid(True)
plt.title('Mass on a Spring (No Damping)')
plt.legend()
plt.show()



#plot damped - just displacement
plt.plot(solutionset.t, solutionset.y[0], color = 'red')
plt.title('Mass on a spring (with damping)')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.grid(True)
plt.show()





#plot damped and non damped together to show difference - just displacement

plt.plot(solution.t, solution.y[0], color = 'black', label='No damping')
plt.plot(solutionset.t, solutionset.y[0], color = 'red', label = 'Damping')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.legend()
plt.grid(True)
plt.show()


```
