```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wt = np.linspace(0,2*np.pi, 1000)

Ah =  eval(input('enter horizontal maginitude'))    #horizontal amplitude
phaseh = eval(input('enter horizontal phase (0<phi<2pi)'))    #horizontal phase

Av = eval(input('enter vertical maginitude'))         #verical amplitude
phasev = eval(input('enter vertical phase (0<phi<2pi)'))     #vertical phase

wp = eval(input('enter waveplate value, or 0 for no waveplate'))    #waveplate


def Eh(Ah, phaseh, wt):       #horizontal component
    return Ah*np.cos(wt + phaseh)

def Ev(Av, phasev, wt):                       #vertical component
    return Av*np.cos(wt + phasev)

plt.plot(wt,Eh(Ah, phaseh, wt))    #plot horizontal
plt.grid('True')
plt.title('Horizontal component')
plt.show()

plt.plot(wt,Ev(Av, phasev, wt))      #plot vertical
plt.grid('True')
plt.title('Vertical component')
plt.show()

Ex = Eh(Ah, phaseh, wt)        #polarization
Ey = Ev(Av, phasev, wt)


plt.plot(Ex,Ey)                 #plotting polarization
plt.title('Polarization pattern')
plt.grid('True')
plt.show()

Eywp = Av*np.cos(wt +phasev+wp)     #vertical component with waveplate

if wp != 0:
  plt.plot(Ex, Eywp)
  plt.title('Polarization with waveplate')     #plot polarization with waveplate
  plt.grid('True')
  plt.show()
```
