```python
#single random walk

import random
import numpy as np
import matplotlib.pyplot as plt

def random_walk():
  N = 10**5
  L = 1
  n= 0
  x, y = 0,0

  x_array = [x]
  y_array = [y]

  while n < N:
    theta = np.random.uniform(0, 2*np.pi)
    dx = L*np.cos(theta)
    dy = L*np.sin(theta)
    x+=dx
    y+=dy
    x_array.append(x)
    y_array.append(y)
    n+=1

  plt.scatter(x_array,y_array, s=1, color = 'b')

random_walk()
```
