```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#assumptions: no air resistance, projectile starts at (0,0)
#dictionary with gravitational constants of different objects in the solar system
celestial_gravity = {
    # Planets
    'Mercury': 3.7,
    'Venus': 8.87,
    'Earth': 9.81,
    'Mars': 3.71,
    'Jupiter': 24.79,
    'Saturn': 10.44,
    'Uranus': 8.69,
    'Neptune': 11.15,

    # Dwarf Planets
    'Pluto': 0.62,
    'Ceres': 0.27,
    'Eris': 0.82,
    'Haumea': 0.44,
    'Makemake': 0.5,

    # Moons
    'Moon': 1.62,
    'Europa': 1.31,
    'Ganymede': 1.43,
    'Callisto': 1.24,
    'Io': 1.79,
    'Titan': 1.35,
    'Enceladus': 0.11,
    'Triton': 0.78,

    # Asteroids
    'Vesta': 0.25,
    'Pallas': 0.23,

    # The Sun
    'Sun': 274.0  # gravity at surface
}

```

```python


#inputs
try:
    v0 = float(input("Enter initial velocity (m/s): "))
    theta_deg = float(input("Enter launch angle in degrees: "))
    celestial_body = input("What celestial body do you want to simulate on: ").capitalize()

    # Check if the celestial body exists in  dictionary
    if celestial_body not in celestial_gravity:
        raise KeyError(f"{celestial_body} is not a valid celestial body")

    theta = np.radians(theta_deg)   #convert angle from degrees to radians
    g = celestial_gravity[celestial_body]

except ValueError:
    print("Please enter a number")
except KeyError as e:
    print(e)
    print("Valid celestial bodies are:", list(celestial_gravity.keys()))  #show valid options
```

```python
#suvat

T= 2*v0*np.sin(theta)/g    #total flight time

t = np.linspace(0, T, num=500)   #time array

x = v0*np.cos(theta)*t  #horizontal displacement

y = (v0*np.sin(theta)*t) - (0.5*g*t**2)  #vertical displacement

y_max = (np.max(y))   #finding maximum vertical height reached numerically (without differentiation), to the nearest whole number
x_max = (np.max(x))
T_1 = (T)
```

```python
#plot graph and information

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.axis("equal")
plt.ylim(0, None)
plt.title(f"Projectile motion, initial velocity={v0}m/s, launch angle={theta_deg}, on {celestial_body}")
plt.xlabel("Horizontal distance travelled (m)")
plt.ylabel("Vertical distance travelled (m)")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.show()

print(f"Total flight time = {T_1}s")
print(f"Maximum vertical displacement = {y_max}m")
print(f"Horizontal distance travelled = {x_max}m")
print(f"g = {g}m/s^2")
```
